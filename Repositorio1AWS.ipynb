```python
nombre_pelicula = 'Talentos Ocultos'
anno = 2016
Protagonista = 'Taraji P. Henson'
Visto = True

# Códigos de color ANSI
rojo = "\033[31m"
verde = "\033[32m"
azul = "\033[34m"
reset = "\033[0m"

print(f"{azul}La Pelicula:{reset} {verde}{nombre_pelicula}{reset} "
      f"{azul}se estreno:{reset} {rojo}{anno}{reset} "
      f"{azul}y su protagonista fue:{reset} {verde}{Protagonista}{reset} "
      f"{azul}visto:{reset} {rojo}{Visto}{reset}")
```

```python
nombres = [
    "The Shawshank Redemption",
    "The Dark Knight",
    "Inception",
    "Pulp Fiction",
    "The Matrix",
    "Interstellar",
    "Parasite",
    "Gladiator",
    "Titanic",
    "The Lord of the Rings: The Fellowship of the Ring"
]

annos = [
    1994, 2008, 2010, 1994, 1999,
    2014, 2019, 2000, 1997, 2001
]

vistas = [
    True, False, True, False, True,
    False, True, False, True, False
]

for nombre, anno, vista in zip(nombres, annos, vistas):
    print(f"La pelicula:{nombre} estreno:{anno} vista:{vista}")
```
