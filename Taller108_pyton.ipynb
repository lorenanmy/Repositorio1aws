```python
import csv
import copy

# Plantilla base del vehículo
myVehicle = {
    "vin": "",
    "make": "",
    "model": "",
    "year": "",
    "range": "",
    "topSpeed": "",
    "zeroSixty": "",
    "mileage": ""
}

# Lista donde se guardarán los vehículos
myInventoryList = []

with open(r'C:\Users\Toshiba\Documents\Lorena\Lorena\Personal Lorena\car_fleet.csv') as csvFile:
    csvReader = csv.reader(csvFile, delimiter=',')
    lineCount = 0

    for row in csvReader:
        if lineCount == 0:
            print(f'Column names are: {", ".join(row)}')
            lineCount += 1
        else:
            print(f'vin: {row[0]} make: {row[1]}, model: {row[2]}, year: {row[3]}, '
                  f'range: {row[4]}, topSpeed: {row[5]}, zeroSixty: {row[6]}, mileage: {row[7]}')

            currentVehicle = copy.deepcopy(myVehicle)

            currentVehicle["vin"] = row[0]
            currentVehicle["make"] = row[1]
            currentVehicle["model"] = row[2]
            currentVehicle["year"] = row[3]
            currentVehicle["range"] = row[4]
            currentVehicle["topSpeed"] = row[5]
            currentVehicle["zeroSixty"] = row[6]
            currentVehicle["mileage"] = row[7]

            myInventoryList.append(currentVehicle)
            lineCount += 1

    print(f'Processed {lineCount} lines.')

print("\nInventario cargado:")
print(myInventoryList)
```
