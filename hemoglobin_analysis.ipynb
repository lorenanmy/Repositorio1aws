```python
"""
hemoglobin_analysis.py

Script para analizar la secuencia de proteína de la hemoglobina beta humana.

Funciones del script:
1. Leer la secuencia desde un archivo de texto.
2. Mostrar información básica de la secuencia.
3. Calcular la composición de aminoácidos.
4. Calcular el peso molecular de la proteína.
5. Calcular el porcentaje de aminoácidos hidrofóbicos.
6. Guardar los resultados en un archivo JSON.
"""

import json

# ----------------------------------------
# Leer la secuencia desde archivo
# ----------------------------------------

with open("hemoglobin_clean.txt", "r") as file: ## abre el archivo en modo lectura
    sequence = file.read().strip()  ## Elimina los espacios al inicio y al final, quita los saltos de linea

# ----------------------------------------
# Mostrar información básica
# ----------------------------------------

print("Protein name: Human Hemoglobin Beta")
print("Sequence length:", len(sequence)) ## Cuenta cuántos aminoácidos tiene la proteína
print("Sequence preview:", sequence[:30], "...") ## sequence[:30] indica que toma los primeros 30 digitos

# ----------------------------------------
# Lista de aminoácidos
# ----------------------------------------

amino_acids = [
'A','R','N','D','C','E','Q','G',
'H','I','L','K','M','F','P','S',
'T','W','Y','V'
]   ## La lista contiene los 20 aminoácidos estándar que forman las proteínas

# ----------------------------------------
# Conteo de aminoácidos
# ----------------------------------------

aa_counts = {}

for aa in amino_acids:
    aa_counts[aa] = sequence.count(aa)

print("\nAmino acid composition:")
for aa, count in aa_counts.items():
    print(aa, ":", count)

# ----------------------------------------
# Pesos moleculares de aminoácidos
# ----------------------------------------

aa_weights = {
'A': 89.09,
'R': 174.20,
'N': 132.12,
'D': 133.10,
'C': 121.15,
'E': 147.13,
'Q': 146.15,
'G': 75.07,
'H': 155.16,
'I': 131.17,
'L': 131.17,
'K': 146.19,
'M': 149.21,
'F': 165.19,
'P': 115.13,
'S': 105.09,
'T': 119.12,
'W': 204.23,
'Y': 181.19,
'V': 117.15
}

# ----------------------------------------
# Función para calcular peso molecular
# ----------------------------------------

def calculate_molecular_weight(sequence):

    total_weight = 0

    for aa in sequence:
        if aa in aa_weights:
            total_weight += aa_weights[aa]

    return total_weight


molecular_weight = calculate_molecular_weight(sequence)

print("\nMolecular weight:", molecular_weight)

# ----------------------------------------
# Porcentaje de aminoácidos hidrofóbicos
# ----------------------------------------

hydrophobic = ['A','V','I','L','M','F','W','Y']

hydrophobic_count = 0

for aa in sequence:
    if aa in hydrophobic:
        hydrophobic_count += 1

hydrophobic_percentage = (hydrophobic_count / len(sequence)) * 100

print("\nHydrophobic amino acids:", hydrophobic_count)
print("Hydrophobic percentage:", round(hydrophobic_percentage,2), "%")

# ----------------------------------------
# Guardar resultados en JSON
# ----------------------------------------

results = {
    "protein_name": "Human Hemoglobin Beta",
    "sequence_length": len(sequence),
    "amino_acid_counts": aa_counts,
    "molecular_weight": molecular_weight,
    "hydrophobic_percentage": hydrophobic_percentage
}

with open("hemoglobin_results.json", "w") as json_file:
    json.dump(results, json_file, indent=4)

print("\nResults saved to hemoglobin_results.json")
```
