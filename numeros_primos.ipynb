```python
def es_primo(n):
    if n < 2:
        return False
    for i in range(2, n):
        if n % i == 0:
            return False
    return True

primos = []

for numero in range(1, 251):
    if es_primo(numero):
        primos.append(numero)

# Mostrar en pantalla
print("Números primos entre 1 y 250:")
for p in primos:
    print(p)

# Guardar en archivo
with open("results.txt", "w") as archivo:
    for p in primos:
        archivo.write(str(p) + "\n")

print("Los resultados fueron guardados en results.txt")
```
